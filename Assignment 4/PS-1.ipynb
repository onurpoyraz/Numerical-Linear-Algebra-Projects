# Exercise 4.4
Two matrices $A,B \in C^{m \times m}$ are **unitarily equivalent** if $A = QBQ^*$ for some unitary $Q \in C^{m \times m}$. Is it true or false that $A$ and $B$ are unitarily equivalent if and only if they have the same singular values?

## Solution
First we decompose $A$ using SVD:
\begin{align}
    A &= U_A \Sigma_A V_A^*     
\end{align}

Multiplying $A$ with unitary matricex $U$ and $V$, we get the following:
\begin{align}
    U_A^* A V_A &= U_A^*U_A \Sigma_A V_A^*V_A    
\end{align}

Since $U^* U = I$, we can write it as follows:
\begin{align}
    U_A^* A V_A &= \Sigma_A    
\end{align}

Now, we can write $A$ as $QBQ^*$:
\begin{align}
    U_A^* Q B Q^* V_A &= \Sigma_A    
\end{align}

If we also decompose $B$ using SVD, the equation turns to the following:
\begin{align}
    U_A^* Q U_B \Sigma_B V_B^*  Q^* V_A &= \Sigma_A     
\end{align}

We know that the multiplication of unitary matrices is unitary too. Since multiplication by unitary matrices does not affect the magnitude of the singular values, singular values of $A$ and $B$ are same.
\begin{align}    
    (U_A^* Q U_B) \Sigma_B (V_B^* Q^* V_A) &= \Sigma_A \\
\end{align}



    

#Exercise 5.1
In Example 3.1 we considered the matrix (3.7) and asserted, among other things, that its 2-norm is approximately 2.9208. Using the SVD, work out (on paper) the exact values of $\sigma_{min}(A)$ and $\sigma_{max}(A)$ for the matrix. 

Note: A is given in (3.7) as follows:
$$ A = \left[
    \begin{array}{cc}
      1&2\\
      0&2
    \end{array}
\right] $$

## Solution
We know that $\Vert{A}\Vert_2$ is equal to the square root of largest eigenvalue of $A^T A$.

\begin{align}
    A &= U \Sigma V^* \Rightarrow U^* A V = \Sigma    
\end{align}

\begin{align}
    \Vert{U^* A V}\Vert_2 &= \Vert{\Sigma}\Vert_2 \\
    \Vert{U^* A V}\Vert_2 &= \Vert{A V}\Vert_2 \\
     &= \Vert{A}\Vert_2 = \Vert{\Sigma}\Vert_2
\end{align}

\begin{align}
    \Vert{A}\Vert_2 &= \Vert{\Sigma}\Vert_2 = \sup\limits_{\Vert{x}\Vert_2 = 1}{\Vert{\Sigma x}\Vert_2}=\sigma_1
\end{align}

Also, we can use the following relationship:
\begin{align}
    \big{|}\ det(A)\ \big{|} &= \big{|}\ det(U \Sigma V^*\ \big{|} \\
    &= \big{|}\ det(U)\ \big{|} \big{|}\ det(\Sigma)\ \big{|} \big{|}\ det(V^*)\ \big{|} \\
    &= \big{|}\ det(\Sigma)\ \big{|} =  \sigma_1 . \sigma_2
\end{align}
Since we have already know that $\big{|}\ det(A)\ \big{|} = 1\times2 - 0\times0 = 2$, and $\sigma_1 = 2.9208$, we can say that $\sigma_2 = 2/2.9208 = 0.6847$.

#Exercise 5.5
Suppose $A \in \mathbb{C}^{m\times m}$ has an SVD $A = U \Sigma V^*$. Find an eigenvalue decomposition (5.1) of the $2m \times 2m$ hermitian matrix
$$ B = \left[
    \begin{array}{cc}
      0&A^*\\
      A&0
    \end{array}
\right] $$.

## Solution
\begin{align}
    A &= U \Sigma V^* \Rightarrow A V = U \Sigma \\
    A^* &= V \Sigma^* U^* = V \Sigma U^* \Rightarrow A^* U = V \Sigma
\end{align}

We know that eigenvalues can be calculated as follows:
\begin{align}
det(B - \lambda I) = 0 \\
\left|
    \begin{array}{cc}
      -\lambda I&A^*\\
      A&-\lambda I
    \end{array}
\right| = 0\\
\lambda^2 I  - A^* A = 0 \\
\lambda^2 I - V \Sigma^* U^* U \Sigma V^* = 0 \\
| \lambda^2 I - V \Sigma^* \Sigma V^* | = |0| \\
| V^* \lambda^2 I V - \Sigma^* \Sigma | = 0 \\
| \lambda^2 I V^* V - \Sigma^* \Sigma | = 0 \\
| \lambda^2 I - \Sigma^2 | = 0 
\end{align}
So $\lambda = \pm \Sigma$ and we can write eigenvalue decomposition as:
\begin{align}
B &= Q \left[
    \begin{array}{cc}
      \Sigma & 0\\
      0 & -\Sigma
    \end{array}
\right] Q^* \\ 
B Q &= Q \left[
    \begin{array}{cc}
      \Sigma & 0\\
      0 & -\Sigma
    \end{array}
\right] \\
\left[
    \begin{array}{cc}
      0&A^*\\
      A&0
    \end{array}
\right]  \left[
    \begin{array}{cc}
      c_{11} & c_{12}\\
      c_{21} & c_{22}
    \end{array}
\right] &= \left[
    \begin{array}{cc}
      c_{11} & c_{12}\\
      c_{21} & c_{22}
    \end{array}
\right] \left[
    \begin{array}{cc}
      \Sigma & 0\\
      0 & -\Sigma
    \end{array}
\right] \\
\left[
    \begin{array}{cc}
      A^*c_{21} & A^*c_{22}\\
      Ac_{11} & Ac_{12}
    \end{array}
\right] &= \left[
    \begin{array}{cc}
      c_{11} \Sigma & -c_{12}\Sigma\\
      c_{21} \Sigma & -c_{22}\Sigma
    \end{array}
\right] 
\end{align}
Since we know that $A V = U \Sigma$ and $A^* U = V \Sigma$, we can say that eigenvector are:
\begin{align}
C = \left[
    \begin{array}{cc}
      V & -V\\
      U & U
    \end{array}
\right] 
\end{align}

Then we need to normalize $C$ such that it is a unitary matrix. 

Therefore, the eigenvalue decomposition can be written as:
\begin{align}
B = (\dfrac{1}{\sqrt{2}}\left[
    \begin{array}{cc}
      V & -V\\
      U & U
    \end{array}
\right]) \left[
    \begin{array}{cc}
      \Sigma & 0\\
      0 & -\Sigma
    \end{array}
\right] (\dfrac{1}{\sqrt{2}}\left[
    \begin{array}{cc}
      V^* & U^*\\
      -V^* & U^*
    \end{array}
\right]) 
\end{align}
