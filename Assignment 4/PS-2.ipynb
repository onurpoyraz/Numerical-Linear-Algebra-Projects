# Exercise 6.1
If $P$ is an orthogonal projector, then $I-2P$ is unitary. Prove this algebraically, and give a geometric interpretation.
## Solution
\begin{align}
    P\  is\  orthogonal\  projector &\iff P = P^*\  and\  P = P ^2 \\
    U\  is\  unitary\  &\iff U^* U = I
\end{align}

For a orthognoal projector $P$, we need to show the following for $I-2P$ to be unitary: 
\begin{align}
    (I-2P)^* (I-2P) = I
\end{align}

Using the properties of $P$, we can can conclude that $I-2P$ is unitary:
\begin{align}
    (I-2P)^* (I-2P) &= (I-2P^*) (I-2P)  \\
    &= I-2P-2P^*+4P^*P \\
    &= I-2P-2P+4PP \\
    &= I-4P+4P \\
    &= I \\
\end{align}

__Geometric Interpretation__:

$I-2P$ substracts $2Px$ from $x$ and reflects $x$ with respect to range of $P$.
    

#Exercise 6.5
Let $P \in \mathbb{C}^{m\times m}$ be a nonzero projector. Show that $\Vert{P}\Vert_2 \geq 1$, with equality if and only if P is an orthogonal projector.

## Solution
First we will show that:
\begin{align}
P\  is\  an\  orthogonal\  projector\  \Rightarrow\ \Vert{P}\Vert_2 \geq 1
\end{align}
\begin{align}
    \Vert{x}\Vert_2^2 &= x^* x \\
    &= x^*Px + x^*(I-P)x \\
    &= x^*PPx + x^*(I-P)(I-P)x \\
    &= x^*P^*Px + x^*(I-P)^*(I-P)x \\
    &= (Px)^*Px + ((I-P)x)^*(I-P)x \\
    &= \Vert{Px}\Vert_2^2  + \Vert{(I-P)x}\Vert_2^2 \\
    &\geq \Vert{Px}\Vert_2^2
\end{align}
Therefore,
\begin{align}
    \Vert{x}\Vert_2^2 &\geq \Vert{Px}\Vert_2^2 \\
    \Vert{x}\Vert_2 &\geq \Vert{Px}\Vert_2 \\
    1 &\geq \dfrac{\Vert{Px}\Vert_2}{\Vert{x}\Vert_2}
\end{align}
\begin{align}
    \Vert{P}\Vert_2  &= \sup\limits_{\Vert{x}\Vert_2 \neq 0}{\dfrac{\Vert{Px}\Vert_2}{\Vert{x}\Vert_2}} \\
    &=1
\end{align}

We also need to show that:
\begin{align}
    \Vert{P}\Vert_2 = 1 \Rightarrow\ P\  is\  an\  orthogonal\  projector\  
\end{align}

Lemma:
\begin{align}
    PP^*PP^*  &= PP^* + ((P-I)P^*)^* ((P-I)P^*) \\
\end{align}
Proof:
\begin{align}
    PP^*PP^*  &= PP^* + ((P-I)P^*)^* ((P-I)P^*) \\
              &= PP^* + (P(P-I)^*) ((P-I)P^*) \\
              &= PP^* + (PP^*-P) (PP^*-P^*) \\
              &= PP^* + PP^*PP^* - PP^*P^* - PPP^* + PP^* \\
              &= PP^* + PP^*PP^* - PP^* - PP^* + PP^* \\              
              &=  PP^*PP^*
\end{align}

So, 
\begin{align}
    x^*PP^*PP^*x &= x^*PP^*x + x^*((P-I)P^*)^* ((P-I)P^*) x \\
    \Vert{PP^*x}\Vert_2^{2} &= \Vert{P^*x}\Vert_2^{2} + \Vert{(P-I)P^*x}\Vert_2^{2}    
\end{align}
Let $y = P^*x$ for any $x \neq 0$
\begin{align}    
    \Vert{Py}\Vert_2^{2} &= \Vert{y}\Vert_2^{2} + \Vert{(P-I)y}\Vert_2^{2}   \ \ (1)\\
    \Vert{y}\Vert_2^{2} &\leq \Vert{Py}\Vert_2^{2} \\
    \Vert{y}\Vert_2 &\leq \Vert{Py}\Vert_2    \ \ (2)
\end{align}
We also know that:
\begin{align}    
    \Vert{P}\Vert_2 &\geq \dfrac{\Vert{Py}\Vert_2}{\Vert{y}\Vert_2}  \\
    \Vert{y}\Vert_2 &\geq \Vert{Py}\Vert_2 \ (\  \ \Vert{P}\Vert_2 = 1\ ) \ \ (3)
\end{align}

Considering equations (2) and (3), we can see that the following is true:
\begin{align}    
    \Vert{y}\Vert_2 &= \Vert{Py}\Vert_2 \\
\end{align}

For equation (1) to be hold the following must be true:
\begin{align}    
    \Vert{(P-I)y}\Vert_2 &= 0 \\
    (P-I)y &= 0 \\ 
    (P-I)P^*x &= 0\  \ (\ \ y = P^*x )\\
    (P-I)P^* &= 0 \ \ (\ \ x \neq 0\ )\\
    PP^* -P^* &= 0 \ \\
    PP^* &= P^*   \  \ (4)
\end{align}
By taking conjugate transpose we get:
\begin{align}
    PP^* &= P \ \ (5) \\
\end{align}
Combining equations (4) and (5), we can see that $P = P^*$ and $P$ is an orthogonal projector.
